```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the Fire Incidents CSV file
fire_df = pd.read_csv("Fire Incidents.csv")

# Print the columns (for reference)
print("Fire Incidents CSV columns:", fire_df.columns.tolist())

# 2. Preprocess the 'neighborhood' column: convert to string, strip spaces, and convert to lower-case
fire_df['neighborhood'] = fire_df['neighborhood'].astype(str).str.strip().str.lower()

# 3. Aggregate fire incident counts by neighborhood
fire_counts = fire_df['neighborhood'].value_counts().sort_values(ascending=False).reset_index()
fire_counts.columns = ['neighborhood', 'fire_count']

# Print top 10 neighborhoods by fire incidents (for reference)
print("\nTop 10 Neighborhoods by Fire Incident Count:")
print(fire_counts.head(10))

# 4. Extract the top 10 and bottom 10 neighborhoods from the aggregated results
top10 = fire_counts.head(10)
bottom10 = fire_counts.tail(10)  # bottom 10 (lowest fire counts)

# 5. Create the graphs for top 10 and bottom 10 using subplots
fig, axs = plt.subplots(2, 1, figsize=(12, 12))

# Graph for the Top 10 Neighborhoods
axs[0].bar(top10['neighborhood'], top10['fire_count'], color='red')
axs[0].set_title("Top 10 Neighborhoods by Fire Incident Count in Pittsburgh")
axs[0].set_xlabel("Neighborhood")
axs[0].set_ylabel("Fire Incident Count")
axs[0].tick_params(axis='x', rotation=45)

# Graph for the Bottom 10 Neighborhoods
axs[1].bar(bottom10['neighborhood'], bottom10['fire_count'], color='blue')
axs[1].set_title("Bottom 10 Neighborhoods by Fire Incident Count in Pittsburgh")
axs[1].set_xlabel("Neighborhood")
axs[1].set_ylabel("Fire Incident Count")
axs[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()


```
