```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# 1. CSV 파일 읽기
df = pd.read_csv("Police Zone.csv")

# 2. zone 컬럼의 값들이 몇 번씩 등장하는지 계산
zone_counts = df["zone"].value_counts()

# 3. 바 차트로 시각화
plt.figure(figsize=(10, 5))
zone_counts.plot(kind="bar")          # 막대그래프
plt.xlabel("Zone")                    # X축 라벨
plt.ylabel("Count")                   # Y축 라벨
plt.title("Number of Rows per Zone")  # 그래프 제목
plt.show()
```
